# Определение перспективного тарифа для телеком-компании

## Описание проекта:

Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.  

Необходимо сделать предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

**Описание тарифов:**

**Тариф «Смарт»:**  

1. Ежемесячная плата: 550 рублей;
2. Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика;
3. Стоимость услуг сверх тарифного пакета:  
   - минута разговора: 3 рубля;  
   - сообщение: 3 рубля;  
   - 1 Гб интернет-трафика: 200 рублей.

**Тариф «Ультра»:**  

1. Ежемесячная плата: 1950 рублей;
2. Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика;
3. Стоимость услуг сверх тарифного пакета:  
   - минута разговора: 1 рубль;  
   - сообщение: 1 рубль;  
   - 1 Гб интернет-трафика: 150 рублей.

**Описание данных:**

Таблица *users* (информация о пользователях):  
* `user_id` — уникальный идентификатор пользователя
* `first_name` — имя пользователя
* `last_name` — фамилия пользователя
* `age` — возраст пользователя (годы)
* `reg_date` — дата подключения тарифа (день, месяц, год)
* `churn_date` — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
* `city` — город проживания пользователя
* `tariff` — название тарифного плана  

Таблица *calls* (информация о звонках):
* `id` — уникальный номер звонка
* `call_date` — дата звонка
* `duration` — длительность звонка в минутах
* `user_id` — идентификатор пользователя, сделавшего звонок  

Таблица *messages* (информация о сообщениях):
* `id` — уникальный номер сообщения
* `message_date` — дата сообщения
* `user_id` — идентификатор пользователя, отправившего сообщение

Таблица *internet* (информация об интернет-сессиях):
* `id` — уникальный номер сессии
* `mb_used` — объём потраченного за сессию интернет-трафика (в мегабайтах)
* `session_date` — дата интернет-сессии
u* `ser_id` — идентификатор пользователя

Таблица *tariffs* (информация о тарифах):
* `tariff_name` — название тарифа
* `rub_monthly_fee` — ежемесячная абонентская плата в рублях
* `minutes_included` — количество минут разговора в месяц, включённых в абонентскую плату
* `messages_included` — количество сообщений в месяц, включённых в абонентскую плату
* `mb_per_month_included` — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
* `rub_per_minute` — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
* `rub_per_message` — стоимость отправки сообщения сверх тарифного пакета
* `rub_per_gb` — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

**Примечания:**  

«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута. 

Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта. Примите, что неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.

## Расчеты и добавление данных в таблицу

### Количество сделанных звонков и израсходованных минут

Рассчитаем количество сделанных звонков и израсходованных минут разговора каждого пользователя по месяцам:

```python
users_calls = calls.groupby(['user_id', calls['call_date'].dt.month])['duration'].agg(['count','sum']).reset_index()
```

```python
users_calls.head()
```

Переименуем столбцы:

```python
users_calls = users_calls.rename(columns={'sum':'sum_duration', 'count':'count_calls', 'call_date':'month'})
```

```python
users_calls.head()
```

### Количество отправленных сообщений

Рассчитаем количество отправленных сообщений по месяцам:

```python
users_messages = messages.groupby(['user_id', messages['message_date'].dt.month])['id'].agg('count').reset_index()
```

```python
users_messages.head()
```

Переименуем столбцы:

```python
users_messages = users_messages.rename(columns={'id':'count_messages', 'message_date':'month'})
```

```python
users_messages.head()
```

### Объем израсходованного интернет-трафика

Рассчитаем объема израсходованного интернет-трафика по месяцам:

```python
users_internet = internet.groupby(['user_id', internet['session_date'].dt.month])['mb_used'].agg('sum').reset_index()
```

```python
users_internet.head()
```

Переименуем столбцы:

```python
users_internet = users_internet.rename(columns={'session_date':'month'})
```

```python
users_internet.head()
```

### Помесячная выручка с каждого пользователя

Создадим общую таблицу, которая объединит таблицы *users_calls*, *users_messages*, *users_internet*.

Изучим таблицы, которые будем объединять:

```python
users_calls.info()
```

```python
users_messages.info()
```

```python
users_internet.info()
```

В качестве основной таблицы для при присоединения выберем таблицу с большим количеством данных - таблица *users_internet*.  
Присоединим к ней таблицу *users_calls* по столбцам 'user_id' и 'month':

```python
users_internet_calls= users_internet.merge(users_calls, on=['user_id', 'month'], how='outer')
```

```python
users_internet_calls.info()
```

К получившейся таблице присоединим таблицу *users_messages* так же по столбцам 'user_id', 'month':

```python
users_general = users_internet_calls.merge(users_messages, on=['user_id', 'month'], how='outer')
```

```python
users_general.info()
```

```python
users_general.head()
```

Для анализа нам потребуются данные о городе пользователя и тарифе, который он использует.  
Возьмем эти данные из таблицы *users*:

```python
users_tariff = users[['user_id', 'tariff', 'city']]
```

Добавим эти данные в общую таблицу *users_general*:

```python
users_general = users_general.merge(users_tariff, on='user_id', how='left')
```

```python
users_general.head()
```

Добавим в общую таблицу информацию об условиях тарифов:

```python
users_general = users_general.merge(tariffs, left_on = 'tariff', right_on = 'tariff_name', how='left')
```

```python
users_general.head()
```

Удалим лишний столбец *tariff_name*:

```python
users_general = users_general.drop(columns='tariff_name')
```

```python
users_general.info()
```

В получившейся таблице имеются пропуски в столбцах 'mb_used', 'count_calls', 'sum_duration' и 'count_messages'. Во всех этих столбцах содержится информация о том, в каком количестве абонент пользовался теми или иными услугами. Примем, что пропуски означают, что абонент не пользовался услугой. В таком случае их можно заменить на 0.

```python
users_general = users_general.fillna(0)
```

Добавим столбец с выручкой за минуты разговора:

```python
users_general['revenue_minutes'] = (users_general['sum_duration'] - users_general['minutes_included'])*users_general['rub_per_minute']
```

В условиях сказано, что "Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта."  
Поэтому переведем Мб в Гб и округлим в большую сторону.  
Добавим столбец с выручкой за интернет трафик:

```python
users_general['revenue_mb']= ((users_general['mb_used'] - users_general['mb_per_month_included'])/1024).apply(np.ceil)*users_general['rub_per_gb']
```

Добавим столбец с выручкой за сообщения:

```python
users_general['revenue_messages'] = (users_general['count_messages'] - users_general['messages_included'])*users_general['rub_per_message']
```

## Проверка гипотез

### Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

Сформулируем нулевую гипотезу H0:  
*Средняя выручка пользователей тарифом «Смарт» равна средней выручке пользователей тарифом «Ультра»*

Сформулируем альтернативную гипотезу H1:  
*Средняя выручка пользователей тарифом «Смарт» отличается от средней выручки пользователей тарифом «Ультра»*

Найдем выручки для пользователей тарифом «Смарт»:

```python
revenue_smart = users_general[users_general['tariff'] =='smart']['revenue']
```

Найдем выручки для пользователей тарифом «Ультра»:

```python
revenue_ultra = users_general[users_general['tariff'] =='ultra']['revenue']
```

Так как мы имеем дело лишь с информацией из небольшой выборки пользователей, то мы не можем просто сравнить средние. Нам необходимо воспользоваться статистическим методом - t-тестом.

Проведем t-тест:

```python
alpha = 0.05

results = st.ttest_ind(revenue_smart, revenue_ultra)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Промежуточный вывод:**  
p-значение очень мало и это показывает, что вероятность того, что H0 верна очень низкая и поэтому мы отвергаем нулевую гипотезу.  
И можем сделать вывод:  
*Средняя выручка пользователей тарифом «Смарт» отличается от средней выручки пользователей тарифом «Ультра»*.

### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Сформулируем нулевую гипотезу H0:  
*Средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов*

Сформулируем альтернативную гипотезу H1:  
*Средняя выручка пользователей из Москвы отличается от средней выручки пользователей из других регионов*

Найдем выручки для пользователей из Москвы:

```python
revenue_moscow = users_general[users_general['city'] == 'Москва']['revenue']
```

Найдем выручки для пользователей из других регионов:

```python
revenue_regions = users_general[users_general['city'] != 'Москва']['revenue']
```

Так как мы имеем дело лишь с информацией из небольшой выборки пользователей, то мы не можем просто сравнить средние. Нам необходимо воспользоваться статистическим методом - t-тестом.

Проведем t-тест:

```python
alpha = 0.05

results = st.ttest_ind(revenue_moscow, revenue_regions)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Промежуточный вывод:**  
p-значение 0,59 показывает, что вероятность того, что H0 верна составляет примерно 59%. Поэтому мы не можем отвергнуть нулевую гипотезу и можем сделать вывод:  
*Средняя выручка пользователей из Москвы примерно равна средней выручке пользователей из других регионов*
